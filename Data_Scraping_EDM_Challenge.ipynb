```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import csv
```

###1. Inisialisasi variable yang dibutuhkan
beberapa hal yang kita butuhkan adalah :


*   Variable url sebagai sumber
*   Response sebagai permintaan request terhadap url

*   Soup sebagai objek dari response yang bisa dilakukan operasi scraping





```python
url = 'https://aksi.bangda.kemendagri.go.id/emonev/DashPrev'
response = requests.get(url, verify=False)
soup = BeautifulSoup(response.text, 'html.parser')
```

### 2.Lakukan pengambilan data berdasarkan element html

Rapihkan data ke dalam format yang ingin digunakan untuk pemrosesan lebih lanjut (List of lists)

```python
data = []

for i in range(1, 7):
  url = f'https://aksi.bangda.kemendagri.go.id/emonev/DashPrev/index/{i}'
  response = requests.get(url, verify=False)
  soup = BeautifulSoup(response.text, 'html.parser')
  data_tahun = []

  for index, row in enumerate(soup.find_all('tr')):
    row_data = []
    if index <= 1:
      continue

    for cell in row.find_all('td'):
      row_data.append(cell.text)
    data_tahun.append(row_data)
    row_data = []
  data.append(data_tahun)
  data_tahun = []

data # data sudah berbentuk list of lists yang berisikin data masing masing daerah
```

###Tampilkand data dalam bentuk dataframe untuk pemrosesan lebih lanjut

```python
df_2019 = pd.DataFrame(data[0])
df_2020 = pd.DataFrame(data[1])
df_2021 = pd.DataFrame(data[2])
df_2022 = pd.DataFrame(data[3])
df_2023 = pd.DataFrame(data[4])
df_2024 = pd.DataFrame(data[5])

dfs = [df_2019, df_2020, df_2021, df_2022, df_2023, df_2024]

print('2019\n')
print(df_2019)
print('2020\n')
print(df_2020)
print('2021\n')
print(df_2021)
print('2022\n')
print(df_2022)
print('2023\n')
print(df_2023)
print('2024\n')
print(df_2024)
```

### 3.Lakukan data cleansing

Hapus kolom yang tidak relevan dan berikan juga nama category sebagai judul masing-masing column.

Hapus juga row terakhir karena bukan data yang diinginkan

```python
def clean_dfs(df):
    df = df.drop(df.columns[0], axis=1)
    df = df.iloc[:-1]

    return df

dfs_cleaned = [clean_dfs(df) for df in dfs]

# Mengembalikan nilai ke masing-masing variabel df
df_2019, df_2020, df_2021, df_2022, df_2023, df_2024 = dfs_cleaned

print('2019\n')
print(df_2019)
print('2020\n')
print(df_2020)
print('2021\n')
print(df_2021)
print('2022\n')
print(df_2022)
print('2023\n')
print(df_2023)
print('2024\n')
print(df_2024)
```

```python
def add_column_names(df):
    column = ['provinsi', 'jumlah balita', 'stunting (pendek)', 'stunting (sangat pendek)', 'persentase stunting']
    df.columns = column
    return df

dfs_added_column = [add_column_names(df) for df in dfs_cleaned]

# Mengembalikan nilai ke masing-masing variabel df
df_2019, df_2020, df_2021, df_2022, df_2023, df_2024 = dfs_added_column

print('2019\n')
print(df_2019)
print('2020\n')
print(df_2020)
print('2021\n')
print(df_2021)
print('2022\n')
print(df_2022)
print('2023\n')
print(df_2023)
print('2024\n')
print(df_2024)
```

#### 4.Konversi tipe data menjadi tipe data yang sesuai
Kita akan membuat kolom Jumlah balita, stunting (pendek), dan stunting (sangat pendek) menjadi int.

Sedangkan untuk persentase kita akan konversi menjadi float.

```python
# membuat fungsi untuk konvesi df

def convert_columns(df):
    # Konversi kolom ke int
    df['jumlah balita'] = df['jumlah balita'].str.replace(',', '').astype(int)
    df['stunting (pendek)'] = df['stunting (pendek)'].str.replace(',', '').astype(int)
    df['stunting (sangat pendek)'] = df['stunting (sangat pendek)'].str.replace(',', '').astype(int)

    # Konversi kolom ke float
    df['persentase stunting'] = df['persentase stunting'].str.replace('%', '').astype(float)

    return df

# Looping fungsi ke masing masing df
dfs_converted = [convert_columns(df) for df in dfs_added_column]

# Mengembalikan nilai ke masing-masing variabel df
df_2019, df_2020, df_2021, df_2022, df_2023, df_2024 = dfs_converted
```

### 5.Simpan semua dfs menjadi csv

```python
# Simpan semua DataFrame dalam sebuah dictionary
dfs = {
    'df_2019': df_2019,
    'df_2020': df_2020,
    'df_2021': df_2021,
    'df_2022': df_2022,
    'df_2023': df_2023,
    'df_2024': df_2024
}

# Menyimpan setiap DataFrame sebagai CSV
for name, df in dfs.items():
    df.to_csv(f'{name}.csv', index=False)
```
